## IMPORTING ALL THE NECESSARY LIABRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
df=pd.read_csv('D:\Blue Bash\modified.csv')
```

```python
df.head()
```

```python
df.info()
```

### APPLYING TF-IDF TO THE 'review/text' FEATURE/COLUMN

```python
text_feature = 'review/text'
target_feature = 'review/overall'
numerical_features = ['review/taste', 'review/palate', 'review/aroma', 
                      'beer/name_encoded', 'beer/style_encoded']
```

```python
# Apply TF-IDF vectorization to 'review/text'
tfidf = TfidfVectorizer(max_features=100, ngram_range=(1, 2), stop_words='english')
text_tfidf = tfidf.fit_transform(df[text_feature].fillna(""))
```

```python
# Convert TF-IDF matrix to DataFrame
text_tfidf_df = pd.DataFrame(text_tfidf.toarray(), columns=tfidf.get_feature_names_out())
```

### SPLITTING THE DATA INTO X AND y AS BELOW:
1. X ---> Independent Features
2. y ---> Target Feature

```python
# Combine TF-IDF features with numerical features
X = pd.concat([df[numerical_features].reset_index(drop=True), text_tfidf_df], axis=1)
y = df[target_feature]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

### RANDOM FOREST MODEL

```python
# Train a Random Forest Regressor
model = RandomForestRegressor(n_estimators=100, random_state=42)  # You can tune hyperparameters
model.fit(X_train, y_train)
```

```python
# Predict and evaluate
y_pred = model.predict(X_test)
```

```python
# Calculate metrics
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Display metrics
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"R² Score: {r2:.4f}")
```

### LINEAR REGRESSION MODEL

```python
# Train a regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict and evaluate
_pred = model.predict(X_test)
# Calculate metrics
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Display metrics
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"R² Score: {r2:.4f}")
```

```python
## THIS IS THE SCORE OF R2 VALUE
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
```

```python
## THIS IS THE SCORE OF ADJUSTED R2 VALUE
adjusted_r2 = 1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

```python
print("The slope or coefficient of weight is ",model.coef_)  # THIS IS THE VALUE OF THETA-1
print("Intercept:",model.intercept_) ## THIS IS THE VALUE OF THETA-0
```
